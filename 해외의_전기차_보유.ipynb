```python
# prompt: /content/Electric_Vehicle_Population_Data.csv 를 읽고 각 city 마다 어떤 차를 많이 가지고 있는지 표로 나타내고 옆에 따로 저 데이터의 내용을 정리하는 차트를 보여줘

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')

# 도시별로 가장 많이 보유한 차량 모델 찾기
city_top_models = df.groupby(['City'])['Make'].agg(lambda x: x.mode()).to_frame()
city_top_models.rename(columns={'Make':'Most Frequent Make'}, inplace=True)

# 도시별로 가장 많이 보유한 차량 모델과 개수 출력
city_model_counts = df.groupby(['City', 'Make'])['Make'].count().unstack(fill_value=0)
city_model_counts['Most Frequent Make'] = city_top_models['Most Frequent Make']

# 표 출력
print(city_model_counts)


# 추가 차트 생성
# 예시: 도시별 전기차 모델 분포 차트
# 전기차 모델 종류별 데이터 정리
make_counts = df.groupby('Make')['Make'].count().sort_values(ascending=False)

# 차트 시각화
plt.figure(figsize=(10, 6))
plt.bar(make_counts.index, make_counts.values)
plt.xticks(rotation=90)
plt.xlabel('전기차 모델')
plt.ylabel('개수')
plt.title('도시별 전기차 모델 분포')
plt.tight_layout()
plt.show()
```
