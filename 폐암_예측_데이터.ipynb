#*표본집단 소개*

```python
# prompt: /content/Lung Cancer Dataset.csv 에서 AGE 랑 GENDER 에 해당하는 값들을 막대그래프로 나타내줘

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일을 Pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/Lung Cancer Dataset.csv')

# 'AGE'와 'GENDER' 열을 선택합니다.
age_gender_df = df[['AGE', 'GENDER']]

# 'GENDER'별 'AGE'의 분포를 막대 그래프로 나타냅니다.
plt.figure(figsize=(10, 6))
for gender in age_gender_df['GENDER'].unique():
  plt.hist(age_gender_df[age_gender_df['GENDER'] == gender]['AGE'], bins=10, alpha=0.5, label=gender)

plt.xlabel('AGE')
plt.ylabel('Frequency')
plt.title('Distribution of Age by Gender')
plt.legend()
plt.show()
```

#*흡연의 유무와 폐암의 관계*

```python
# prompt: /content/Lung Cancer Dataset.csv 에서 SMOKING 에 따른 PULMONARY_DISEASE 의 값을 표로 나타내줘

import pandas as pd

# CSV 파일을 Pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/Lung Cancer Dataset.csv')

# 'SMOKING'과 'PULMONARY_DISEASE' 열을 선택합니다.
smoking_pulmonary_df = df[['SMOKING', 'PULMONARY_DISEASE']]

# 'SMOKING'에 따른 'PULMONARY_DISEASE'의 값을 교차표로 나타냅니다.
pd.crosstab(smoking_pulmonary_df['SMOKING'], smoking_pulmonary_df['PULMONARY_DISEASE'])
```

#*전조증상(초기증상)과의 관계*

```python
# prompt: /content/Lung Cancer Dataset.csv 에서 FINGER_DISCOLORATION  , BREATHING_ISSUE , THROAT_DISCOMFORT , CHEST_TIGHTNESS 들과 PULMONARY_DISEASE 의 관계를 각각 표로 나타내줘

import pandas as pd

# CSV 파일을 Pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/Lung Cancer Dataset.csv')

# 분석할 열 목록
columns_to_analyze = ['FINGER_DISCOLORATION', 'BREATHING_ISSUE', 'THROAT_DISCOMFORT', 'CHEST_TIGHTNESS']

# 각 열과 PULMONARY_DISEASE의 관계를 교차표로 나타냅니다.
for column in columns_to_analyze:
  print(f"\n{column}과 PULMONARY_DISEASE의 관계:")
  print(pd.crosstab(df[column], df['PULMONARY_DISEASE']))
```
